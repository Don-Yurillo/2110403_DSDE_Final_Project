```python

# df = df[df['subject_code'] >= 2200]
# df = df[df['subject_code'] <= 2216]
df['subject_code'] = df['subject_code'].apply(lambda x: str(x)[0:2]+str("00"))
df
```

```python
df['predict_gender'] = df['predict_gender'].replace(['m','f'], ['male', 'female'])
df
```

```python
df = pd.DataFrame(df, columns=["subject_code", "predict_gender"])
```

```python
df = df.groupby(['subject_code', 'predict_gender']).value_counts().reset_index(name='count')
df.columns = ['subject_code', 'gender', 'count(sub)']
df
```

```python
total_female = df[df['gender'] == "female"]['count(sub)'].sum()
total_female
```

```python
total_male =  df[df['gender'] == "male"]['count(sub)'].sum()
total_male
```
